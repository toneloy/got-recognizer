```python
import requests

got_characters_wikipedia_url = 'https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_characters'
got_characters_wikipedia = requests.get(got_characters_wikipedia_url).text
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import re

soup = BeautifulSoup(got_characters_wikipedia, 'html')
table = soup.find('table', {'class': 'wikitable'})
rows = table.find_all('tr')

search_terms = []
for row in rows:
    row_tds = row.find_all('td')
    if len(row_tds) > 1:
        actor_td = row_tds[0]
        actor_a = actor_td.find('a')
        actor = actor_a.get('title')
        
        character_td = row_tds[1]
        character_a = character_td.find('a')
        

        character = character_a.get('title')
        if character:
            
            character = re.sub('\(.*?\)', '', character).strip().lower()
            actor = re.sub('\(.*?\)', '', actor).strip().lower()
            
            search_terms.append("{} {}".format(character, actor))

search_terms
```

```python
from google_images_download import google_images_download

response = google_images_download.googleimagesdownload()
absolute_image_paths = [response.download({'keywords': search_term, 'limit': 10}) for search_term in search_terms]
```
